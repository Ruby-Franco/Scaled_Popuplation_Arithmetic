```python
import pandas as pd 
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("results.csv")
```

```python
df.head(5)
```

```python
df = df.rename(columns={'Absolute': 'Absolute Error'})
df.head(5)
```

```python
df.hist(column="Absolute Error",bins=20,  color =  u'#348ABD')
```

```python
df['% Difference'] = (df['Absolute Error']/df['Actual'])*100
```

```python
df.head(1000)
```

```python
df['% Difference'].replace(np.inf, 0, inplace=True)
df.head(1000)
```

```python
df.hist(column='% Difference',bins=20, color =  u'#A60628');
```

```python
df['% Difference 2'] = (df['% Difference']/100)
```

```python
x = df["Absolute Error"]
y = df['% Difference 2']
bins = np.linspace(0, 1, 20)
plt.style.use('bmh')

plt.hist([x, y], bins, label=['Absolute Error', '% Difference'])
plt.legend(loc='upper right')
plt.show()
```
